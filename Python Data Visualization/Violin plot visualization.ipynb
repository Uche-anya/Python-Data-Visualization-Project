```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + np.random.normal(0, 0.2, 100)
y2 = np.cos(x) + np.random.normal(0, 0.2, 100)
 
# Violin Plot 
plt.figure(figsize=(8, 6))
plt.violinplot([y1, y2], showmedians=True)
plt.xticks([1, 2], ['Sin(x) + Noise', 'Cos(x) + Noise'])
plt.title('Violin Plot - Random Data Exploration')
plt.ylabel('Value')
plt.show()
```
