```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + np.random.normal(0, 0.2, 100)
y2 = np.cos(x) + np.random.normal(0, 0.2, 100)

# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='Sin(x) + Noise')
plt.plot(x, y2, label='Cos(x) + Noise')
plt.title('Line Plot - Random Data Exploration')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
