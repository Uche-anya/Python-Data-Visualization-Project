```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + np.random.normal(0, 0.2, 100)
y2 = np.cos(x) + np.random.normal(0, 0.2, 100)

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(y1, bins=20, alpha=0.7, label='Sin(x) + Noise', color='blue')
plt.hist(y2, bins=20, alpha=0.7, label='Cos(x) + Noise', color='orange')
plt.title('Histogram - Random Data Exploration')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()


```
